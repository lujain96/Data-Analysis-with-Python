```python
### Open Your CSV file and print the first 10 rows in a good format ###
import pandas as pd
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head(10)
```

```python
### Check if any NANs in your dataset and fill them with a good filer ###
movies.isnull().sum()
movies["homepage"].fillna("No homepage", inplace = True)
movies["overview"].fillna("No overview", inplace = True)
movies["release_date"].fillna((movies["release_date"].mode()[0]), inplace = True)
movies["runtime"].fillna((movies["runtime"].mean()), inplace = True)
movies["tagline"].fillna("No tagline", inplace = True)
movies.isnull().sum()
```

```python
### Discover the types of each columns and modify them if needed ###
movies.dtypes
```

```python
### Give a very simple statistical analsys for the numerical columns ###
movies.describe()
```

```python
### Calculate the mean rate for [1999, 1980, 2004, 2017] the movies grouped by years ###

### Ex: Average Movies Rate at 2009 is 6.065587 ###
l = []
movies['year'] = pd.DatetimeIndex(movies['release_date']).year
grouped_movies = movies.groupby('year')
grouped_movies_mean = grouped_movies['vote_average'].mean()
print('Average Movies Rate at 1999 is:',grouped_movies_mean.loc[1999])
print('Average Movies Rate at 1980 is:',grouped_movies_mean.loc[1980])
print('Average Movies Rate at 2004 is:',grouped_movies_mean.loc[2004])
print('Average Movies Rate at 2017 is:',grouped_movies_mean.loc[2017])
```

```python
### Rearange the dataframe based on revenue - budget values ###
movies["earnings"] = movies["revenue"] - movies["budget"]
movies.sort_values(by="earnings", ascending=False, inplace=True)
movies
```

```python
### Allocate the year with the largest movies released in ['Action', 'Romance'] ###
import ast
def convert(obj):
    lst = []
    for i in ast.literal_eval(obj):
        lst.append(i['name'])
    return lst

movies['genres'] = movies['genres'].apply(convert)
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
movies[movies.genres == "['Action','Romance']"]
```

```python
### Find the movies with top 5 revenue - budget value ###
movies.head()
```

```python
### Find the year with the highest number of movies release ###
high = movies['year'].value_counts()
high
print(high.index[0])
```

```python
### Find the top 2 countries with the highest production movies number ###
def convert2(obj):
    lst = []
    for i in ast.literal_eval(obj):
        lst.append(i['iso_3166_1'])
    return lst

movies['production_countries'] = movies['production_countries'].apply(convert2)

country = movies['production_countries'].value_counts()
print(country.index[0], country.index[1])
```

```python
### Find the top 1 company with the highest production movies number ###
movies['production_companies'] = movies['production_companies'].apply(convert)
com = movies['production_companies'].value_counts()
print(com.index[1])
#movies['production_companies']
```
