```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sales_df = pd.read_csv('Data/company_sales_data.csv')
sales_df.head()

# x axis values
month = sales_df['month_number']
# corresponding y axis values
profit = sales_df['total_profit']

# plotting the points
plt.plot(month, profit)
# setting x and y axis range
plt.xticks(month)
plt.yticks(np.arange(100000, 600000, step=100000))
# naming the x axis
plt.xlabel('Month number')
# naming the y axis
plt.ylabel('Profit in dollar')

# giving a title to the graph
plt.title('Company profit per month')

# function to show the plot
plt.show()
```

```python
sales_df.plot.line(x = 'month_number', y = 'total_profit', title='Company profit per month')
plt.xticks(month)
plt.yticks(np.arange(100000, 600000, step=100000))
# naming the x axis
plt.xlabel('Month number')
# naming the y axis
plt.ylabel('Profit in dollar')
```

```python
sns.lineplot(data=sales_df, x = 'month_number', y = 'total_profit').set_title('Company profit per month')
plt.xticks(month)
plt.yticks(np.arange(100000, 600000, step=100000))
# naming the x axis
plt.xlabel('Month number')
# naming the y axis
plt.ylabel('Profit in dollar')
```

#### 2. Read all product sales data and display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Note:**It must be implemented in three methods  
The graph should look like this.


```python
# x axis values
month = sales_df['month_number']
# corresponding y axis values
faceCream = sales_df['facecream']
faceWash = sales_df['facewash']
toothPaste = sales_df['toothpaste']
bathingSoap = sales_df['bathingsoap']
shampoo = sales_df['shampoo']
moisturizer = sales_df['moisturizer']

# plotting the points
plt.plot(month, faceCream, label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month, faceWash, label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(month, toothPaste, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, bathingSoap, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, shampoo, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, moisturizer, label = 'ToothPaste Sales Data', marker='o', linewidth=3)

# setting x and y axis range
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
# naming the x axis
plt.xlabel('Month Number')
# naming the y axis
plt.ylabel('Sales units in number')
# specifying the legend location
plt.legend(loc='upper left')
# giving a title to the graph
plt.title('Sales data')
# show the plot
plt.show()
```

```python
sns.lineplot(markers=True, data=sales_df.drop(['total_units', 'total_profit'], axis=1)).set_title('Sales data')
# setting x and y axis range
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
# specifying the legend location
plt.legend(loc='upper left')
# naming the x axis
plt.xlabel('Month Number')
# naming the y axis
plt.ylabel('Sales units in number')
```

```python
sales_df.plot.line(x = 'month_number', y = ['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo','moisturizer'] , style='o-', title = 'Sales data')
# setting x and y axis range
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
# naming the x axis
plt.xlabel('Month Number')
# naming the y axis
plt.ylabel('Sales units in number')
```

#### 3. Read face cream and facewash product sales data and show it using the bar chart.


- The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.  
- **Note:**It must be implemented in three methods  

The bar chart should look like this.


```python
a4_dims = (8,8)
x_ind=np.arange(len(month))
x=0.25
fig, ax = plt.subplots(figsize=a4_dims)
plt.bar(x_ind-x,sales_df['facecream'],width=x,label="Face cream sales data")
plt.bar(x_ind, sales_df['facewash'], width=x,label="Face wash sales data")
# givin a title 
plt.title("Average of Lessons")
# naming the x axis
plt.xlabel("Month Number")
# naming the y axis
plt.ylabel("Sales units in number")
# specifying the legend location
plt.legend(loc ='upper left')
plt.grid(linestyle='--', linewidth=1)
plt.xticks(ticks=x_ind, labels=month)
```

```python
fig, ax1 = plt.subplots(figsize=(8,8))
temp = sales_df.drop(['toothpaste', 'bathingsoap','shampoo','moisturizer','total_units', 'total_profit'], axis=1)
sales = temp.melt(id_vars='month_number').rename(columns=str.title)
labels = sales_df[['facecream', 'facewash']]
sns.barplot(x=sales_df['month_number'], y = 'Value', hue = 'Variable', data=sales, ax=ax1)
sns.despine(fig)
plt.grid(linestyle='--', linewidth=1)
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
```

```python
sales_df.plot(x="month_number", y=["facecream", "facewash"], kind="bar")
plt.grid(linestyle='--', linewidth=1)
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
```

#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.  
The histogram should look like this.


```python
# create figure and axis
a4_dims = (8,8)
fig, ax = plt.subplots(figsize=a4_dims)
profitList = sales_df['total_profit'].tolist()
labels = ['low', 'average', 'Good', 'Best']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
# naming the x and y axis
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
# specifiying the legent location
plt.legend(loc='upper left')
plt.xticks(profit_range)
# giving a title
plt.title('Profit data')
plt.show()
```

```python
sales_df.plot.hist(profitList, profit_range, label = 'Profit data', color= 'blue', title = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
a4_dims = (8,8)
fig, ax = plt.subplots(figsize=a4_dims)
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'BathingSoap', 'Shampoo', 'Moisturizer']
sales = [sales_df['facecream'].sum(), sales_df['facewash'].sum(), sales_df['toothpaste'].sum(), 
         sales_df['bathingsoap'].sum(), sales_df['shampoo'].sum(), sales_df['moisturizer'].sum()]
plt.axis("equal")
plt.pie(sales, labels=labels, autopct='%1.1f%%')
# legend location
plt.legend(loc='lower right')
# giving a title
plt.title('Sales data')
plt.show()
```

```python
a4_dims = (8,8)
fig, ax = plt.subplots(figsize=a4_dims)
sales = sales_df.drop(['total_profit', 'total_units'], axis = 1)
plt.axis("equal")
axes = sales.sum().plot.pie(autopct="%.1f%%");
ax.set_title('Sales data')
```
